# Bayes theorem
$P(H|E) = \frac{P(H)P(E|H)}{P(E)} = \frac{P(H)P(E|H)}{P(H)P(E|H) + P(\neg H)P(E|\neg H)} = \frac{nP(H)P(E|H)}{nP(H)P(E|H) + nP(\neg H)P(E|\neg H)}$

Стив тихий и застенчивый человек. Стив библиотекарь или фермер?
Соотношение библиотекарей к фермерам 1 к 20, это значит на 10 библиотекарей приходится 200 фермеров.
4 из 10 библиотекарей и 20 из 200 фермеров тихие и застенчивые.  
Это значит что вероятность что стив библиотекарь $\frac{4}{4 + 20} \approx 16.7\%$  
Даже если вы считаете что библиотекари тихони в четыре раза чаще, фермеры легко компенсируют это своим количеством.

## Представим всё в формуле
У нас есть гипотеза: мы считаем что Стив библиотекарь  
Тут нам дают описание где он оказывается тихим и застенчивым  
$P(Hypothesis|Evidence)$ 
Эта черта(|) означает "при условии что" иными словами нас теперь интересует только варианты попадающие под описание Стива (доля).

$n$ = популяция, в нашем случае 210  
$P(H)$ = доля библиотекарей 1/21  
$P(E|H)$ = доля тихих и застенчивых среди библиотекарей 0.4  
$P(\neg H)$ = доля фермеров 20/21  
$P(E|\neg H)$ = доля тихих и застенчивых среди фермеров 0.1   

**Основаня идея теоремы Байеса** состоит в том чтобы получить большую точность оценки вероятности события путём учёта дополнительных данных. То есть, новые данные не должны формировать ваши убеждения с чистого листа, они лишь уточняют старые.  
В нашем случае новыми данными можно считать что Стив тихий и застенчивый.

**Принцп прост:** есть первоначальная основная оценка вероятности, которую уточняют с получением большого количества информации.

## [Пример](https://neurohive.io/ru/osnovy-data-science/kak-primenjat-teoremu-bajesa-dlja-reshenija-realnyh-zadach/)
Давайте рассмотрим теорему Байеса более подробно на примере со сном. Есть данные за два месяца, показывающие во сколько человек засыпает. Например если мы возьмём время 22:00, то вероятность что человек уже спит будет равна 27.34%.  
Итак, зададим вопрос, как изменится вероятность что человек спит в 22:00, если учесть время, в течение которого включен свет в спальне? Для уточнения оценки и нужна теорема Байеса.

$P(sleep|light) = \frac{P(light|sleep)P(sleep)}{P(light)}$

Выражение сверху - вероятность того, что человек спит, при условии, что известно, включён ли свет в спальне.   
Из собственных наблюдений нам известно следущее: вероятность того, что человек спит, когда свет включен, равна 1%. 

$P(light|sleep) = 0.01$  вероятность что свет **включен** во время сна
$P(-light|sleep) = 0.99$  вероятность что свет **выключен** во время сна  

Уравнение также включает в себя константу нормировки $P(light)$ - вероятность того что свет включен. Свет бывает включен и когда человек спит, и когда он бодрствует. Поэтому, зная априорную вероятность сна, вычислим константу нормировки так: 

$P(light) = P(light|sleep) * P(sleep) + P(light|-sleep) * P(-sleep)$   

Вероятность того, что свет включен в тот момент, когда человек не спит, равна $P(light|-sleep)$ и определяется путём наблюдения. Нам известно что свет горит, когда человек бодрствует с вероятностью 80% (это значит, что есть 20% вероятность того, что свет не включен, если человек бодрствует).

$P(light|-sleep) = 0.8$

Применим формулу к моменту времени 22:30 и учтём, что свет включен. Мы знаем, вероятность того, что человек спал, равна 73.90%. Это число - отправная точка для нашей оценки. Уточним его, учтя информацию об освещении. Зная, что свет включен, подставим числа в формулу:

$P(sleep|light) = \frac{0.7390 * 0.01}{0.01 * 0.7390 + 0.8 * (1 - 0.7390)} = 0.0342$

Дополнительные данные резко изменили оценку вероятности: от более 70% до 3.42%. Это показывает силу теоремы Байеса: мы смогли уточнить нашу первоначальную оценку ситуации, включив в неё больше информации.

## Добавим больше данных
Например, мы знаем, что если телефон заряжается, то человек спит с вероятностью 95%. Этот факт можно учесть в нашей модели. Предположим, что вероятность того, что телефон заряжается, не зависит от освещения в спальне (независимость событий - это достаточно сильное упрощение, но оно позвоит сильно облегчить задачу). Составим новое, ещё более точное выражение для верности:

$P(sleep|light, phone) = \frac{P(light|sleep)P(phone|sleep)P(sleep)}{P(light|sleep)P(phone|sleep)P(sleep) + P(light|-sleep)P(phone|-sleep)P(-sleep)}$

Вычислив вероятность для 23:00 (95.52% вероятность что человек спит) с учётом что свет горит, а телефон не заряжается мы получим $1.74%$  
В 23:00 без дополнительной информации мы могли почти с полной вероятностью сказать, что человек спит. Однако, как только у нас будет дополнительная информация о том, что свет включен, а телефон не заряжается, мы заключаем что вероятность того, что человек спит, практически равна нулю.
